```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Courses/ML A-Z Udemy/Clustering/Mall_Customers.csv')
df
```

```python
X = df.iloc[:,[3,4]].values
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

###	â€¢	K-Means: Partitions data into k clusters by minimizing the variance within each cluster.
###	â€¢	Hierarchical Clustering: Builds a hierarchy of clusters using either a top-down (divisive) or bottom-up (agglomerative) approach.
###	â€¢	DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Groups points based on density, effectively identifying clusters of varying shapes and handling noise.

```python
# Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Apply Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=4)
hierarchical_labels = hierarchical.fit_predict(X)

# Apply DBSCAN Clustering
dbscan = DBSCAN(eps=1.0, min_samples=3)
dbscan_labels = dbscan.fit_predict(X)
```

```python
def plot_clusters(X, labels, title):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', s=50)
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
plot_clusters(X, kmeans_labels, "K-Means Clustering")
```

```python
plot_clusters(X, hierarchical_labels, "Hierarchical Clustering")
```

```python
plt.figure(figsize=(10, 5))
linked = linkage(X, method='ward')
dendrogram(linked)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
plot_clusters(X, dbscan_labels, "DBSCAN Clustering")
```
